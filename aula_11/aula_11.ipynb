```python
from threading import Thread
from time import time, sleep
```

```python
def fat(n):
    if n <= 1:
        return n
    elif n == 2:
        return 1
    else:
        return fat(n-1) + fat(n-2)
        
```

```python
def aux(n,resultados,i):
    x = fat(n)
    resultados[i] = str(x)
```

```python
def durma(n):
    print('acordo em breve')
    sleep(n)
    print('acordei')
    return n
def durma2(n):
    print('acordo em breve')
    sleep(n)
    print('acordei')
    return n
```

```python
start = time()
class minhaThread (Thread):
    def __init__(self,my_function,args):
        Thread.__init__(self)
        self.my_function = my_function
        self.args = args
    def run(self):
        if self.my_function == 'durma':
            durma(self.args)
        elif self.my_function == 'durma2':
            durma2(self.args)
for i in range(5):
    threads = []
    #Criando as threads
    thread1 = minhaThread('durma',i)
    thread1.start()
    threads.append(thread1)
    
    thread2 = minhaThread('durma2',i)
    thread2.start()
    threads.append(thread1)
    for t in threads:
        t.join()

print(time() - start)
```

```python


def durma(n,resultados,i):
    print('pessoa 1 dormindo')
    sleep(n)
    print('pessoa 1 acordada')
    resultados[i] = n  * 1
    
def durma2(n,resultados,i):
    print('pessoa 2 dormindo')
    sleep(n)
    print('pessoa 2 acordada')
    resultados[i] = n * 2
    
def durma3(n,resultados,i):
    print('pessoa 3 dormindo')
    sleep(n)
    print('pessoa 3 acordada')
    resultados[i] = n * 3
    
def durma4(n,resultados,i):
    print('pessoa 4 dormindo')
    sleep(n)
    print('pessoa 4 acordada')
    resultados[i] = n * 4
    
def durma5(n,resultados,i):
    print('pessoa 5 dormindo')
    sleep(n)
    print('pessoa 5 acordada')
    resultados[i] = n * 5
    
def durma6(n,resultados,i):
    print('pessoa 6 dormindo')
    sleep(n)
    print('pessoa 6 acordada')
    resultados[i] = n * 6
    
    
    
```

```python
start = time()
class minhaThread (Thread):
    def __init__(self,my_function,arg1,arg2,arg3):
        Thread.__init__(self)
        self.my_function = my_function
        self.arg1 = arg1
        self.arg2 = arg2
        self.arg3 = arg3
    def run(self):
        self.my_function(self.arg1,self.arg2,self.arg3)
resultados1 = [None] * 5
resultados2 = [None] * 5
resultados3 = [None] * 5
resultados4 = [None] * 5
resultados5 = [None] * 5
resultados6 = [None] * 5
for i in range(5):
    threads = []
    #Criando as threads
    thread1 = minhaThread(durma,i,resultados1,i)
    thread1.start()
    threads.append(thread1)
    
    thread2 = minhaThread(durma2,i,resultados2,i)
    thread2.start()
    threads.append(thread2)
    
    thread3 = minhaThread(durma3,i,resultados3,i)
    thread3.start()
    threads.append(thread3)
    
    thread4 = minhaThread(durma4,i,resultados4,i)
    thread4.start()
    threads.append(thread4)
    
    thread5 = minhaThread(durma5,i,resultados5,i)
    thread5.start()
    threads.append(thread5)
    
    thread6 = minhaThread(durma6,i,resultados6,i)
    thread6.start()
    threads.append(thread6)
    
    for t in threads:
        t.join()

print(time() - start)
print(resultados1)
print(resultados2)
print(resultados3)
print(resultados4)
print(resultados5)
print(resultados6)
```
